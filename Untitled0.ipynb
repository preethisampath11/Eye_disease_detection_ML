```python
# Set random seed for reproducibility
tf.random.set_seed(42)

# Constants
IMG_SIZE = 224  # EfficientNetB0 default input size
BATCH_SIZE = 32
EPOCHS = 20

def create_model(num_classes):
    """Create and compile the EfficientNetB0 model."""
    # Load EfficientNetB0 with pre-trained weights
    try:
        base_model = EfficientNetB0(
            weights='imagenet',
            include_top=False,
            input_shape=(IMG_SIZE, IMG_SIZE, 3)
        )
    except Exception as e:
        print(f"Warning: failed to load imagenet weights ({e}); initialising EfficientNetB0 without weights.")
        base_model = EfficientNetB0(
            weights=None,
            include_top=False,
            input_shape=(IMG_SIZE, IMG_SIZE, 3)
        )

    # Freeze the base model layers
    base_model.trainable = False

    # Add custom layers for multi-label classification
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(1024, activation='relu')(x)
    x = Dropout(0.5)(x)
    x = Dense(512, activation='relu')(x)
    x = Dropout(0.3)(x)
    outputs = Dense(num_classes, activation='sigmoid')(x)

    model = Model(inputs=base_model.input, outputs=outputs)

    # Compile the model
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
        loss='binary_crossentropy',
        metrics=['accuracy', tf.keras.metrics.AUC()]
    )

    return model

def create_data_generators():
    """Create data generators for training and validation."""
    train_datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest'
    )

    val_datagen = ImageDataGenerator(rescale=1./255)

    return train_datagen, val_datagen
```

```python
# Load and prepare data
def load_data(csv_path, image_ext='.png'):
    """Load and prepare the dataset labels."""
    df = pd.read_csv(csv_path)

    # Derive label columns: everything after ID and Disease_Risk
    label_columns = df.columns[2:]

    # Build filename column: if ID already has an extension, keep it; otherwise append image_ext
    id_str = df['ID'].astype(str)
    has_ext = id_str.str.contains(r"\.(?:jpg|jpeg|png|bmp|tif|tiff)$", case=False, regex=True)
    df['filename'] = np.where(has_ext, id_str, id_str + image_ext)

    return df, label_columns

# Load training data
train_csv = f'{TRAIN_DIR}/RFMiD_Training_Labels.csv'
val_csv = f'{VAL_DIR}/RFMiD_Validation_Labels.csv'

train_df, label_columns = load_data(train_csv)
val_df, _ = load_data(val_csv)

num_classes = len(label_columns)
print(f"Number of classes: {num_classes}")

# Create model
model = create_model(num_classes)

# Create data generators
train_datagen, val_datagen = create_data_generators()

# Set up data generators - point to the images subdirectory
train_generator = train_datagen.flow_from_dataframe(
    dataframe=train_df,
    directory=TRAIN_IMG_DIR,  # Point to images subfolder
    x_col="filename",
    y_col=label_columns.tolist(),
    class_mode="raw",
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    shuffle=True,
    validate_filenames=False
)

val_generator = val_datagen.flow_from_dataframe(
    dataframe=val_df,
    directory=VAL_IMG_DIR,  # Point to images subfolder
    x_col="filename",
    y_col=label_columns.tolist(),
    class_mode="raw",
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    shuffle=False,
    validate_filenames=False
)
```

```python
# Create callbacks
checkpoint_cb = tf.keras.callbacks.ModelCheckpoint(
    f'{MODEL_PATH}/best_model.h5',
    save_best_only=True,
    monitor='val_loss'
)

early_stopping_cb = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)

# Train the model
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=EPOCHS,
    callbacks=[checkpoint_cb, early_stopping_cb]
)

# Save the final model
model.save(f'{MODEL_PATH}/final_model.h5')

# Save training history
hist_df = pd.DataFrame(history.history)
hist_df.to_csv(f'{HISTORY_PATH}/training_history.csv', index=False)

print(f"Training completed! Models saved to: {MODEL_PATH}")
print(f"Training history saved to: {HISTORY_PATH}")
```

```python
import matplotlib.pyplot as plt

# Plot training history
def plot_training_history(history):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    # Plot accuracy
    ax1.plot(history.history['accuracy'])
    ax1.plot(history.history['val_accuracy'])
    ax1.set_title('Model Accuracy')
    ax1.set_ylabel('Accuracy')
    ax1.set_xlabel('Epoch')
    ax1.legend(['Train', 'Validation'])

    # Plot loss
    ax2.plot(history.history['loss'])
    ax2.plot(history.history['val_loss'])
    ax2.set_title('Model Loss')
    ax2.set_ylabel('Loss')
    ax2.set_xlabel('Epoch')
    ax2.legend(['Train', 'Validation'])

    plt.tight_layout()
    plt.show()

    # Save the plot
    fig.savefig(f'{HISTORY_PATH}/training_history_plot.png')

plot_training_history(history)
```

```python
# Install and import Gradio
!pip install gradio
import gradio as gr
import numpy as np
from PIL import Image
import tensorflow as tf

def process_image(image):
    """Process the uploaded image to match model's requirements"""
    img = Image.fromarray(image)
    img = img.resize((IMG_SIZE, IMG_SIZE))
    img_array = np.array(img)
    img_array = img_array / 255.0
    img_array = np.expand_dims(img_array, axis=0)
    return img_array

def predict_diseases(image):
    """Make predictions on the uploaded image"""
    # Process image
    processed_img = process_image(image)

    # Get predictions
    predictions = model.predict(processed_img)[0]

    # Create results dictionary
    results = {}
    for label, prob in zip(label_columns, predictions):
        if prob >= 0.5:  # Only show diseases with probability >= 50%
            results[label] = f"{prob*100:.2f}%"

    if not results:
        return "No diseases detected (all probabilities below 50%)"

    # Format results
    output = "Detected Diseases:\n"
    for disease, probability in results.items():
        output += f"{disease}: {probability}\n"

    return output

# Create Gradio interface
iface = gr.Interface(
    fn=predict_diseases,
    inputs=gr.Image(),
    outputs=gr.Textbox(label="Predictions"),
    title="Eye Disease Detection",
    description="Upload an eye fundus image to detect possible diseases.",
    examples=[],  # You can add example images here if you have any
    theme="default"
)

# Launch the interface
iface.launch(share=True, debug=True)
```
